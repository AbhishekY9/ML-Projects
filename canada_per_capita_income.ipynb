```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

df=pd.read_csv('canada_per_capita_income.csv')

# plot
plt.plot(df['year'],df['per capita income (US$)'])
plt.scatter(df['year'],df['per capita income (US$)'],s=10)

#Train
reg=linear_model.LinearRegression()
reg.fit(df[['year']],df['per capita income (US$)'])
y_pred=reg.predict(df[['year']])


# plot
plt.plot(df['year'],y_pred)
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.title('PCI vs year')
plt.show()


w=reg.coef_
b=reg.intercept_
print(f'Per capita income in 2024 will be {(w[0]*2024+b)}US$')
```
